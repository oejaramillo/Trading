```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
serie = pd.read_csv(r'datos\we_eurusd.csv')
```

## Variable en niveles

```python
serie['Unnamed: 0'] = pd.to_datetime(serie['Unnamed: 0'])

# Configura el tamaño del gráfico
plt.figure(figsize=(10, 6))

# Traza la serie de tiempo
plt.plot(serie['Unnamed: 0'], serie['4. close'], color='blue', label='Precio de cierre')

# títulos y etiquetas
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre semanal')
plt.legend()

# Sombrear 2014
plt.axvspan('2014-01-01', '2014-12-31', alpha=0.2, color='gray')
x_position = pd.Timestamp('2014-01-01') + pd.DateOffset(months=-6)
plt.text(x_position, 1.63, 'Año 2014', fontsize=12, color='black')

# Sombrear 2008
plt.axvspan('2008-01-01', '2008-12-31', alpha=0.2, color='gray')
x_position = pd.Timestamp('2014-01-01') + pd.DateOffset(months=-77)
plt.text(x_position, 1.63, 'Año 2008', fontsize=12, color='black')

plt.axhline(y=1.2, color='gray', linestyle='--')

# Muestra el gráfico
plt.tight_layout()
plt.show()
```

## Variable en primera diferencia

```python
serie['diff'] = serie['4. close'] - serie['4. close'].shift(1)
```
